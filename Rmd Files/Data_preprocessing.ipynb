```python
import pandas as pd
```

```python
df = pd.read_csv("Cars.csv", index_col="Index")
```

```python
len(df)
```

```python
df.head(5)
```

```python
mx = max(df["Price"])
```

```python
mn = min(df["Price"])
```

```python
threshold = (mx - mn)/10
threshold
```

```python
# df['Price_Class'] = 0
```

```python
price_list = list(df["Price"])
class_list = []
for price in price_list:
    if price >= threshold:
        class_list.append("High-priced")
    elif price < threshold:
        class_list.append("Low-priced")
```

```python
len([elem for elem in class_list if elem == "High-priced"])
# for elem in class_list:
#     if elem == "High-priced":
#         print(elem)
```

```python
df["Price_Class"] = class_list
```
